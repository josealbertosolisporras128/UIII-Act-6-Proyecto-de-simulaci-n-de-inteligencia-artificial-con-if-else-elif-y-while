#### Refinanciamiento

```python
# Bienvenida inicial
print("--- IA de Análisis Financiero: Módulo de Refinanciamiento ---")

# Variable de control para el ciclo interactivo
ejecutar_ia = "si"

# Ciclo while para mantener el programa activo [1, 2]
while ejecutar_ia == "si" or ejecutar_ia == "SI":
    print("\nPor favor, ingrese los datos de su crédito actual:")

    # Captura de datos mediante input [3, 4]
    monto_deuda = float(input("Monto total de la deuda: "))
    tasa_interes_actual = float(input("Tasa de interés anual actual (%): "))
    puntaje_credito = int(input("Su puntaje de crédito (300-850): "))
    tasa_nueva_oferta = float(input("Nueva tasa ofrecida por el banco (%): "))

    # Lógica de IA utilizando operadores aritméticos y relacionales [5-7]
    ahorro_tasa = tasa_interes_actual - tasa_nueva_oferta
    ahorro_anual_estimado = monto_deuda * (ahorro_tasa / 100)

    # Estructura condicional if-elif-else para la toma de decisiones [8, 9]
    if puntaje_credito < 620:
        # Uso de operadores relacionales (<)
        print("IA: Análisis detenido. Su puntaje de crédito es demasiado bajo para calificar.")
        print("Sugerencia: Mejore su historial antes de intentar un refinanciamiento.")

    elif ahorro_tasa > 2.0 and puntaje_credito >= 720:
        # Uso de operadores lógicos (and) y relacionales (>, >=) [10, 11]
        print("IA: ¡Excelente oportunidad detectada!")
        print(f"Su ahorro estimado sería de ${ahorro_anual_estimado} al año.")
        print("Decisión: Se recomienda refinanciar de inmediato.")

    elif ahorro_tasa > 0 and ahorro_tasa <= 2.0:
        # Evaluación de rangos con operadores lógicos [10]
        if ahorro_anual_estimado > 1000:
            print(f"IA: Oportunidad moderada. Aunque la tasa baja poco, su deuda es alta.")
            print(f"Ahorro anual: ${ahorro_anual_estimado}. Decisión: Recomendado.")
        else:
            print("IA: Beneficio insuficiente.")
            print("Decisión: Los costos de cierre podrían ser mayores al ahorro. No recomendado.")

    else:
        # Caso cuando la tasa nueva es mayor o igual a la actual
        print("IA: No hay beneficio económico. La nueva tasa no mejora sus condiciones actuales.")

    # Control de la interactividad del ciclo while [12, 13]
    ejecutar_ia = input("\n¿Desea realizar otro análisis? (si/no): ")

print("\n--- Gracias por usar el Asistente IA de Refinanciamiento ---")
```
